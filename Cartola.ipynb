```python
def procurar_nr_partida(NrsDasPartidas, nrRodada, Id_time):
    # percorre o dicionário NrsDasPartidas
    for item in NrsDasPartidas:
        # verifica se a rodada e o número do time estão presentes na tupla 'times'
        if item['rodada'] == nrRodada and Id_time in item['times']:
            # retorna o valor de 'nr_partida' do dicionário correspondente
            return item['nr_partida']

    # caso não encontre nenhum dicionário que satisfaça as condições, retorna None
    return None
```
